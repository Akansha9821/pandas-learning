## Loc method in pandas are used for acesing the rows and coloumns with their labels

```python
data.loc[10]
```

```python
data.loc[[10,12]]
```

```python
data.loc[10, ["Name", "Team"]]
```

```python
data.loc[data["Number"] < 8, ["Salary"]]
```
